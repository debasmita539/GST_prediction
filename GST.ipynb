```python
#to check whether our dataset is normally distributed or not
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, normaltest, anderson

# Load the dataset
data = pd.read_csv(r'D:\GST\data2.csv')

# List of columns you checked for outliers
columns_to_check = ['Column0',
    'Column1', 'Column2', 'Column3', 'Column4', 'Column5', 'Column6', 'Column7',
    'Column8', 'Column9', 'Column10', 'Column11', 'Column12', 'Column13', 'Column14',
    'Column15', 'Column16', 'Column17', 'Column18', 'Column19', 'Column20', 'Column21'
]

# Function to check normality using statistical tests and plots
def check_normality(data, columns):
    for column in columns:
        # Drop NaN values for the test
        col_data = data[column].dropna()

        # 1. Shapiro-Wilk Test
        stat, p_value = shapiro(col_data)
        print(f"\nShapiro-Wilk Test for {column}:")
        print(f"Test Statistic = {stat}, p-value = {p_value}")
        if p_value > 0.05:
            print(f"{column} appears to be normally distributed (fail to reject H0).")
        else:
            print(f"{column} does not appear to be normally distributed (reject H0).")

        # 2. D'Agostino's K-squared Test
        stat, p_value = normaltest(col_data)
        print(f"\nD'Agostino's K-squared Test for {column}:")
        print(f"Test Statistic = {stat}, p-value = {p_value}")
        if p_value > 0.05:
            print(f"{column} appears to be normally distributed (fail to reject H0).")
        else:
            print(f"{column} does not appear to be normally distributed (reject H0).")

        # 3. Anderson-Darling Test
        result = anderson(col_data)
        print(f"\nAnderson-Darling Test for {column}:")
        print(f"Test Statistic: {result.statistic}")
        for i in range(len(result.critical_values)):
            sl, cv = result.significance_level[i], result.critical_values[i]
            if result.statistic < cv:
                print(f"{column} appears to be normally distributed at the {sl}% level.")
            else:
                print(f"{column} does not appear to be normally distributed at the {sl}% level.")
        
        # 4. Plot Histogram and Q-Q Plot
        plt.figure(figsize=(12, 5))

        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(col_data, bins=30, kde=True)
        plt.title(f'Histogram of {column}')
        
        # Q-Q Plot
        plt.subplot(1, 2, 2)
        from scipy.stats import probplot
        probplot(col_data, dist="norm", plot=plt)
        plt.title(f'Q-Q Plot of {column}')
        
        plt.tight_layout()
        plt.show()

# Run normality checks for specified columns
check_normality(data, columns_to_check)
```

Detect How many OUTLIERS PREESENT ON THE DATASET(ISOLATION FOREST)#Original_Outlier.csv'

```python

import pandas as pd
from sklearn.ensemble import IsolationForest

# Load the dataset
data = pd.read_csv(r'D:\GST\data2.csv')

# List of columns to check for outliers (you can use all columns if needed)
columns_to_check = [
    'Column0', 'Column1', 'Column2', 'Column3','Column4', 'Column5', 'Column6', 
    'Column7', 'Column8', 'Column9', 'Column10', 'Column11', 'Column12', 
    'Column13', 'Column14', 'Column15', 'Column16', 'Column17', 
    'Column18', 'Column19', 'Column20', 'Column21'
]

# Subset the data to include only the columns to check for outliers
subset_data = data[columns_to_check]

# Initialize Isolation Forest
iso_forest = IsolationForest(contamination=0.01, random_state=42)  # Adjust contamination parameter if needed

# Fit the model
iso_forest.fit(subset_data)

# Predict the outliers (returns -1 for outliers and 1 for inliers)
outliers = iso_forest.predict(subset_data)

# Add the outlier flag to the original dataset
data['Outlier_Flag'] = outliers

# Dictionary to store the number of outliers for each column
outliers_per_column = {}

# Count the number of outliers in each column
for column in columns_to_check:
    outliers_per_column[column] = ((data['Outlier_Flag'] == -1) & (data[column].notna())).sum()

# Print the number of outliers detected in each column
print("Number of outliers detected in each column:")
for column, num_outliers in outliers_per_column.items():
    print(f"{column}: {num_outliers} outliers")


# Save the dataset with outlier flags
data.to_csv(r'Original_Outlier.csv', index=False)

print("Dataset with outlier flags saved to 'Original_Outlier.csv")

```

Detect outlier using IQR 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the original dataset
data2 = pd.read_csv(r'D:\GST\data2.csv')

# List of columns to check for outliers
columns_to_check = ['Column0','Column1','Column2', 'Column3','Column4', 'Column5', 'Column6', 'Column7', 'Column8', 'Column9', 'Column10', 'Column11','Column12','Column13','Column14', 'Column15', 'Column16', 'Column17', 'Column18', 'Column19', 'Column20', 'Column21']

# Initialize a dictionary to store the number of outliers in each column
outliers_count = {}

# Loop through each column
for column in columns_to_check:
    # Calculate IQR and bounds for outlier detection
    Q1 = data2[column].quantile(0.25)
    Q3 = data2[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Create a mask to identify rows with outliers
    mask = (data2[column] < lower_bound) | (data2[column] > upper_bound)

    # Count the number of outliers
    outliers_count[column] = mask.sum()

    # Replace outlier values with the median of the column
    median_value = data2[column].median()
    data2.loc[mask, column] = median_value

# Save the cleaned dataset
#data2.to_csv(r'Hey_data_clean.csv', index=False)

# Display the number of outliers detected in each column
print("Number of outliers detected in each column:")
for column, count in outliers_count.items():
    print(f"{column}: {count} outliers")

# Optional: You can also display the total number of outliers
total_outliers = sum(outliers_count.values())
print(f"Total number of outliers detected: {total_outliers}")



```

Find outlier detect using log_transformation approach #outlier_summary_log.csv

```python

import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv(r'D:\GST\data2.csv')

# List of columns to check for outliers (without duplicates)
columns_to_check = [  'Column0','Column1','Column2', 'Column3','Column4', 'Column5', 'Column6', 'Column7', 'Column8', 'Column9', 'Column10','Column11', 'Column12', 'Column13',  'Column14', 'Column15', 'Column16', 'Column17', 'Column18', 'Column19', 'Column20', 'Column21'
]
    


# Function to detect outliers using IQR method after applying log transformation
def detect_outliers_log_iqr(df, column):
    # Apply log transformation (adding 1 to avoid log(0))
    df_log = np.log1p(df[column])
    
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_log.quantile(0.25)
    Q3 = df_log.quantile(0.75)
    IQR = Q3 - Q1  # Interquartile Range (IQR)
    
    # Calculate lower and upper bounds for detecting outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Detect outliers in the transformed data
    outliers = df[(df_log < lower_bound) | (df_log > upper_bound)]
    return outliers

# Dictionary to store the number of outliers for each column
outliers_dict = {}

# Iterate over each column and detect outliers using log transformation and IQR
for column in columns_to_check:
    outliers = detect_outliers_log_iqr(data, column)
    num_outliers = len(outliers)
    outliers_dict[column] = num_outliers

#  number of outliers detected in each column
print("Number of outliers detected in each column (using log transformation):")
for column, num_outliers in outliers_dict.items():
    print(f"{column}: {num_outliers} outliers")
    
total_outliers = sum(outliers_count.values())
print(f"Total number of outliers detected: {total_outliers}")

# Save the results to a CSV file
outliers_summary = pd.DataFrame(list(outliers_dict.items()), columns=['Column', 'Num_Outliers'])
outliers_summary.to_csv('Outliers_Summary_Log.csv', index=False)

print("Outliers summary saved to 'Outliers_Summary_Log.csv'.")

```

Detect outlier using a hybrid approach of(log_trasformation + IQR) Method


```python
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv(r'D:\GST\data2.csv')

# List of columns to check for outliers (without duplicates)
columns_to_check = [  'Column0','Column1','Column2', 'Column3', 'Column5', 'Column6', 'Column7', 'Column8', 'Column9', 'Column10','Column11', 'Column12', 'Column13',  'Column14', 'Column15', 'Column16', 'Column17', 'Column18', 'Column19', 'Column20', 'Column21'
]

# Function to detect outliers using log transformation and IQR method
def detect_outliers_log_iqr(df, column):
    # Apply log transformation (adding 1 to avoid log(0))
    df_log = np.log1p(df[column])
    
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_log.quantile(0.25)
    Q3 = df_log.quantile(0.75)
    IQR = Q3 - Q1  # Interquartile Range (IQR)
    
    # Calculate lower and upper bounds for detecting outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Detect outliers in the original data (not transformed)
    outliers = df[(df[column] < np.exp(lower_bound) - 1) | (df[column] > np.exp(upper_bound) - 1)]
    return outliers

# Dictionary to store the number of outliers for each column
outliers_dict = {}

# Iterate over each column and detect outliers using log transformation and IQR
for column in columns_to_check:
    outliers = detect_outliers_log_iqr(data, column)
    num_outliers = len(outliers)
    outliers_dict[column] = num_outliers

#  number of outliers detected in each column
print("Number of outliers detected in each column (using log transformation and IQR):")
for column, num_outliers in outliers_dict.items():
    print(f"{column}: {num_outliers} outliers")

total_outliers = sum(outliers_dict.values())
print(f"Total number of outliers detected: {total_outliers}")

# Save the results to a CSV file
outliers_summary = pd.DataFrame(list(outliers_dict.items()), columns=['Column', 'Num_Outliers'])
outliers_summary.to_csv('Outliers_Summary_Log_IQR.csv', index=False)

print("Outliers summary saved to 'Outliers_Summary_Log_IQR.csv'.")
```

Replace the outlier Value to Median value  (Log_transformation+IQR)  METHOD  #cleaned_log_transformed_data.csv

```python
import pandas as pd
import numpy as np

# Load the dataset
data2 = pd.read_csv(r'D:\GST\data2.csv')

# List of columns to check for outliers
columns_to_check = ['Column0','Column1','Column2', 'Column3', 'Column4','Column5', 'Column6', 'Column7', 'Column8', 'Column9', 'Column10','Column11','Column12','Column13', 'Column14', 'Column15', 'Column16', 'Column17', 'Column18', 'Column19', 'Column20', 'Column21']  # Add the columns you want to check

# Loop through each column
for column in columns_to_check:
    # Apply log transformation (adding 1 to avoid log(0))
    data2['log_' + column] = np.log(data2[column] + 1)

    # Calculate IQR and bounds for outlier detection on log-transformed data
    Q1 = data2['log_' + column].quantile(0.25)
    Q3 = data2['log_' + column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Create a mask to identify rows with outliers in log-transformed data
    mask = (data2['log_' + column] < lower_bound) | (data2['log_' + column] > upper_bound)

    # Replace outlier values in log-transformed data with the median of the log-transformed column
    median_value_log = data2['log_' + column].median()
    data2.loc[mask, 'log_' + column] = median_value_log

    # Apply inverse log transformation to bring data back to original scale
    data2[column] = np.exp(data2['log_' + column]) - 1

# Save the cleaned dataset
data2.to_csv(r'D:\GST\cleaned_log_transformed_data.csv', index=False)

print("Outliers replaced in log-transformed data, inverse log applied, and data saved to 'cleaned_log_transformed_data.csv'.")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, expr
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.stat import Summarizer

# Initialize Spark Session
spark = SparkSession.builder.appName("DataPreprocessing").getOrCreate()

# Load the dataset
file_path = r'D:\GST\data2.csv'
df = spark.read.csv(file_path, header=True, inferSchema=True)

# List of columns to check for outliers
columns_to_check = ['Column0','Column1','Column2', 'Column3', 'Column4','Column5', 'Column6', 'Column7', 'Column8', 'Column9', 'Column10', 'Column11','Column12','Column13','Column14', 'Column15', 'Column16', 'Column17', 'Column18', 'Column19', 'Column20', 'Column21']  # Add the columns you want to check

# Function to handle outliers using IQR and replace them with the median
def handle_outliers(df, columns):
    for column in columns:
        # Calculate Q1, Q3, and IQR
        Q1 = df.approxQuantile(column, [0.25], 0.05)[0]
        Q3 = df.approxQuantile(column, [0.75], 0.05)[0]
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Create a mask for outliers
        df = df.withColumn(
            column,
            expr(f"CASE WHEN {column} < {lower_bound} OR {column} > {upper_bound} THEN {df.agg({column: 'median'}).collect()[0][0]} ELSE {column} END")
        )
    return df

# Handle outliers in the specified columns
df_cleaned = handle_outliers(df, columns_to_check)

# Save the cleaned dataset
df_cleaned.write.csv(r'D:\GST\cleaned_median_pyspark_data.csv', header=True)

print("Outliers replaced with the median value and data saved to 'cleaned_median_pyspark_data.csv'.")

# Feature Selection
# Create a feature vector for feature selection
feature_columns = [col for col in df.columns if col not in ['Column_to_exclude1', 'Column_to_exclude2']]  # Adjust as needed
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df_assembled = assembler.transform(df_cleaned)

# Standardize features
scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withStd=True, withMean=True)
scaler_model = scaler.fit(df_assembled)
df_scaled = scaler_model.transform(df_assembled)

# Show the cleaned and scaled data
df_scaled.show()

# Stop the Spark session
spark.stop()
```
